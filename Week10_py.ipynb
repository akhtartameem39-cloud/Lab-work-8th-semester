```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_california_housing

housing = fetch_california_housing()
df = pd.DataFrame(housing.data, columns=housing.feature_names)

df['Price'] = housing.target

avg_price_by_age = df.groupby('HouseAge')['Price'].mean()

plt.figure(figsize=(8,5))
plt.plot(avg_price_by_age.index, avg_price_by_age.values, color='blue')
plt.title('Average House Price by House Age')
plt.xlabel('House Age')
plt.ylabel('Average Price')
plt.grid(True)
plt.show()
df['RoomsGroup'] = df['AveRooms'].round()
avg_price_by_rooms = df.groupby('RoomsGroup')['Price'].mean().head(10)

plt.figure(figsize=(8,5))
plt.bar(avg_price_by_rooms.index, avg_price_by_rooms.values, color='green')
plt.title('Average Price by Average Rooms')
plt.xlabel('Average Rooms')
plt.ylabel('Average Price')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
sns.histplot(df['Price'], bins=30, kde=True, color='purple')
plt.title('Distribution of House Prices')
plt.show()

plt.figure(figsize=(8,5))
sns.boxplot(x=df['AveRooms'], color='orange')
plt.title('Boxplot of Average Rooms')
plt.show()

plt.figure(figsize=(10,6))
sns.boxplot(data=df[['MedInc', 'HouseAge', 'AveRooms', 'Population']])
plt.title('Boxplots of Multiple Features')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x=df['MedInc'], y=df['Price'])
plt.title('Median Income vs House Price')
plt.show()

plt.figure(figsize=(6,4))
sns.scatterplot(x=df['AveRooms'], y=df['Price'])
plt.title('Average Rooms vs House Price')
plt.show()

plt.figure(figsize=(10,8))
corr = df.corr()

sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

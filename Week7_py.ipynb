```python
import pandas as pd

# --- Pandas Series from list ---
n = int(input("Enter number of elements for the Series: "))
series_list = [float(input(f"Enter element {i+1}: ")) for i in range(n)]
series = pd.Series(series_list)
print("\nCreated Pandas Series:")
print(series)


# --- DataFrame from dictionary ---
cols = int(input("\nEnter number of columns for DataFrame: "))
data_dict = {}

for i in range(cols):
    col_name = input(f"Enter name of column {i+1}: ")
    values = input(f"Enter values for {col_name} separated by space: ").split()
    values = [float(v) for v in values]
    data_dict[col_name] = values

df_dict = pd.DataFrame(data_dict)
print("\nDataFrame from dictionary:")
print(df_dict)
```

```python
import pandas as pd

# Create DataFrame from user input
rows = int(input("Enter number of rows: "))
cols = int(input("Enter number of columns: "))

data = []
for i in range(rows):
    row = [float(input(f"Enter value for row {i+1}, column {j+1}: ")) for j in range(cols)]
    data.append(row)

df = pd.DataFrame(data)
print("\nDataFrame:")
print(df)


# --- a) Average value of the second column ---
if cols >= 2:
    avg_second_col = df.iloc[:, 1].mean()
    print("\nAverage of second column:", avg_second_col)
else:
    print("\nSecond column not available.")


# --- b) Average of first 5 rows of third and fourth columns ---
if cols >= 4 and rows >= 5:
    avg_3_4 = df.iloc[:5, 2:4].mean()
    print("\nAverage of first 5 rows of 3rd & 4th columns:")
    print(avg_3_4)
else:
    print("\nNot enough rows or columns for this operation.")


# --- c) Row-wise sum ---
row_sum = df.sum(axis=1)
print("\nRow-wise sum:")
print(row_sum)


# --- d) Maximum of average value in each row ---
row_avg = df.mean(axis=1)
max_row_avg = row_avg.max()
print("\nAverage of each row:")
print(row_avg)
print("\nMaximum row average value:", max_row_avg)
```
